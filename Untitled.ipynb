```python
# -----------------------------
# Equations
# -----------------------------

# Power usage limited by capacity
power_usage.equation = sd.min(demand, microgrid_capacity.value)

# Flow: income generated per time step
income_flow.equation = power_usage * cost_per_kwh.value

# Stock: income accumulates the inflow
net_microgrid_income.equation = income_flow

# Maintenance delay improves with higher income
maintenance_delay.equation = sd.lookup(
    net_microgrid_income,
    [(0, 2.0), (5000, 1.6), (15000, 1.0), (30000, 0.7)]
)

# Failure rate depends on maintenance delay
failure_rate.equation = maintenance_delay * 0.05

# SAIFI (interruptions per customer)
saifi.equation = failure_rate / sd.max(1, no_of_customers.value)

# Reliability improves as SAIFI decreases
reliability.equation = sd.max(0.2, 1 - saifi)

# Flow: demand growth depends on reliability
demand_change.equation = demand * baseline_growth_rate.value * reliability

# Stock: demand accumulates the inflow
demand.equation = demand_change

# -----------------------------
# Initial values
# -----------------------------
demand.initial_value = 40000.0
net_microgrid_income.initial_value = 10.0

# -----------------------------
# Constant values
# -----------------------------
average_delay.value = 1
cost_per_kwh.value = 50
microgrid_capacity.value = 1000000
baseline_growth_rate.value = 0.03
no_of_customers.value = 650

# -----------------------------
# Plots
# -----------------------------
print(microgrid_capacity.value)
demand.plot()
net_microgrid_income.plot()
reliability.plot()
saifi.plot()
income_flow.plot()
demand_change.plot()
```

```python
from BPTK_Py import Model
from BPTK_Py import sd_functions as sd
from BPTK_Py import Agent
from BPTK_Py import Model
from BPTK_Py import Event
from BPTK_Py import DataCollector
from BPTK_Py import SimultaneousScheduler
model = Model(starttime=0.0, stoptime=120.0, dt=1.0, name ="Appliance Adoption")

#stocks
tv_ownership_middle = model.stock("tv_ownership_middle")
stove_ownership_middle = model.stock("stove_ownership_middle")
lighting_ownership_middle = model.stock("lighting_ownership_middle")

households_connected = model.stock("households_connected")
households_not_connected = model.stock("households_not_connected")
net_microgrid_income = model.stock("net_microgrid_income")
no_of_failures = model.stock("no_of_failures")
no_of_components = model.stock("no_of_components")
no_of_current_repairs = model.stock("no_of_current_repairs")

net_microgrid_income = model.stock("net_microgrid_income")
no_of_failures = model.stock("no_of_failures")
no_of_components = model.stock("no_of_components")
no_of_current_repairs = model.stock("no_of_current_repairs")

#flows
tv_adoption_middle=model.flow("tv_adoption_middle")
stove_adoption_middle = model.flow("stove_adoption_middle")
lighting_adoption_middle = model.flow("lighting_adoption_middle")
household_microgrid_adoption = model.flow("household_microgrid_adoption")
microgrid_income = model.flow("microgrid_income")
microgrid_expenditures = model.flow("microgrid_expenditures")
microgrid_income_flow = model.flow("microgrid_income_flow")
failure_rate = model.flow("failure_rate")
fixture_rate = model.flow("fixture_rate")

#converters
tv_utility_middle = model.converter("tv_utility_middle")
tv_adoption_prob_middle = model.converter("tv_adoption_prob_middle")
tv_social_influence_middle = model.converter("tv_social_influence_middle")

stove_utility_middle = model.converter("stove_utility_middle")
stove_adoption_prob_middle = model.converter("stove_adoption_prob_middle")
stove_social_influence_middle = model.converter("stove_social_influence_middle")

lighting_utility_middle = model.converter("lighting_utility_middle")
lighting_adoption_prob_middle = model.converter("lighting_adoption_prob_middle")
lighting_social_influence_middle = model.converter("lighting_social_influence_middle")

baseline_demand = model.converter("baseline_demand")
social_influence_microgrid = model.converter("social_influence_microgrid")
tv_demand = model.converter("tv_demand")
stove_demand = model.converter("stove_demand")
lighting_demand = model.converter("lighting_demand")
total_demand = model.converter("total_demand")
power_use = model.converter("power_use")
failure_rate_multiplier =  model.converter("failure_rate_multiplier")
mttr = model.converter("mttr")
saifi = model.converter("saifi")
saidi = model.converter("saidi")
no_of_customers = model.converter("no_of_customers")
downtime = model.converter("downtime")
reliability = model.converter("reliability")
fossil_fuel_baseline_cooking = model.converter("fossil_fuel_baseline_cooking")
fossil_fuel_baseline_lighting = model.converter("fossil_fuel_baseline_lighting")
fossil_fuel_use_cooking = model.converter("fossil_fuel_use_cooking")
fossil_fuel_use_lighting = model.converter("fossil_fuel_use_lighting")

co2_emissions_kerosene = model.converter("co2_emissions_kerosene")
co2_emissions_lpg = model.converter("co2_emissions_lpg")
pm25_emissions_kerosene = model.converter("pm25_emissions_kerosene")
pm25_emissions_lpg = model.converter("pm25_emissions_lpg")
pm25_concentration_daily_kerosene = model.converter("pm25_concentration_daily_kerosene")
pm25_concentration_daily_lpg = model.converter("pm25_concentration_daily_lpg")
pm25_concentration_daily_both = model.converter("pm25_concentration_daily_both")

pm25_exposure_kerosene_men = model.converter("pm25_exposure_kerosene_men")
pm25_exposure_kerosene_women = model.converter("pm25_exposure_kerosene_women")
pm25_exposure_kerosene_children = model.converter("pm25_exposure_kerosene_children")

pm25_exposure_lpg_men = model.converter("pm25_exposure_lpg_men")
pm25_exposure_lpg_women = model.converter("pm25_exposure_lpg_women")
pm25_exposure_lpg_children = model.converter("pm25_exposure_lpg_children")

pm25_exposure_both_men = model.converter("pm25_exposure_both_men")
pm25_exposure_both_women = model.converter("pm25_exposure_both_women")
pm25_exposure_both_children = model.converter("pm25_exposure_both_children")

prob_kerosene = model.converter("prob_kerosene")
prob_lpg = model.converter("prob_lpg")
prob_both = model.converter("prob_both")
prob_stove=model.converter("prob_stove")
prob_lighting=model.converter("prob_lighting")
prob_stove_and_lighting=model.converter("prob_stove_and_lighting")

estimated_cases_kerosene_men = model.converter("estimated_cases_kerosene_men")
estimated_cases_kerosene_women = model.converter("estimated_cases_kerosene_women")
estimated_cases_kerosene_children = model.converter("estimated_cases_kerosene_children")

estimated_cases_lpg_men = model.converter("estimated_cases_lpg_men")
estimated_cases_lpg_women = model.converter("estimated_cases_lpg_women")
estimated_cases_lpg_children = model.converter("estimated_cases_lpg_children")

estimated_cases_both_men = model.converter("estimated_cases_both_men")
estimated_cases_both_women = model.converter("estimated_cases_both_women")
estimated_cases_both_children = model.converter("estimated_cases_both_children")

estimated_cases_men = model.converter("estimated_cases_men")
estimated_cases_women = model.converter("estimated_cases_women")
estimated_cases_children = model.converter("estimated_cases_children")
estimated_cases_total = model.converter("estimated_cases_total")

baseline_cases_both_men = model.converter("baseline_cases_both_men")
baseline_cases_both_women = model.converter("baseline_cases_both_women")
baseline_cases_both_children = model.converter("baseline_cases_both_children")
baseline_cases_total = model.converter("baseline_cases_total")

health_improvements = model.converter("health_improvements")
baseline_lpg_cost = model.converter("baseline_lpg_cost")
stove_cost = model.converter("stove_cost")
cost_savings_lpg = model.converter("cost_savings_lpg")
cost_savings_kerosene = model.converter("cost_savings_kerosene")
baseline_kerosene_cost = model.converter("baseline_kerosene_cost")
lighting_cost = model.converter("lighting_cost")
cost_savings_kerosene = model.converter("cost_savings_kerosene")

cost_savings_both = model.converter("cost_savings_both")

total_baseline=model.converter("total_baseline")
total_electric_cost=model.converter("total_electric_cost")
cost_savings_overall = model.converter("cost_savings_overall")

#constants
alpha_tv_middle = model.constant("alpha_tv_middle") 
beta_income_tv = model.constant("beta_income_tv")
beta_reliability_tv = model.constant("beta_reliability_tv")
beta_social_tv = model.constant("beta_social_tv")
#income = model.constant("income")
initial_tv_composition = model.constant("initial_tv_composition")
tv_utility_sensitivity = model.constant("tv_utility_sensitivity")

alpha_stove_middle = model.constant("alpha_stove_middle") 
beta_income_stove = model.constant("beta_income_stove")
beta_reliability_stove = model.constant("beta_reliability_stove")
beta_social_stove = model.constant("beta_social_stove")
initial_stove_composition = model.constant("initial_stove_compoxition")
stove_utility_sensitivity = model.constant("stove_utility_sensitivity")

alpha_lighting_middle = model.constant("alpha_lighting_middle") 
beta_income_lighting = model.constant("beta_income_lighting")
beta_reliability_lighting = model.constant("beta_reliability_lighting")
beta_social_lighting = model.constant("beta_social_lighting")
initial_lighting_composition = model.constant("initial_lighting_composition")
lighting_utility_sensitivity = model.constant("lighting_utility_sensitivity")

baseline_household_demand = model.constant("baseline_household_demand")
no_of_households = model.constant("no_of_households")
microgrid_capacity = model.constant("microgrid_capacity")
initial_failure_rate = model.constant("initial_failure_rate")
cost_per_kwh = model.constant("cost_per_kwh")
no_of_components_initial_value = model.constant("no_of_components_initial_value")

co2_emissions_factor_kerosene = model.constant("co2_emissions_factor_kerosene")
co2_emissions_factor_lpg = model.constant("co2_emissions_factor_lpg")
pm25_emissions_factor_kerosene = model.constant("pm25_emissions_factor_kerosene")
pm25_emissions_factor_lpg = model.constant("pm25_emissions_factor_lpg")
fraction_of_emissions_kerosene = model.constant("fraction_of_emissions_kerosene")
fraction_of_emissions_lpg = model.constant("fraction_of_emissions_lpg")
air_change_rate = model.constant("air_change_rate")
kitchen_volume = model.constant("kitchen_volume")
incident_rate = model.constant("incident_rate")

lpg_price_per_kg = model.constant("lpg_price_per_kg")
kerosene_price_per_liter = model.constant("kerosene_price_per_liter")


#equations

#tv adoption equations
tv_social_influence_middle.equation = (tv_ownership_middle/no_of_households)*beta_social_tv
tv_utility_middle.equation = (beta_income_tv*income) + (beta_reliability_tv*reliability) + (tv_social_influence_middle)
tv_adoption_prob_middle.equation = 1 / (1 + sd.exp(-(alpha_tv_middle+(tv_utility_sensitivity*tv_utility_middle))))
tv_adoption_middle.equation = tv_adoption_prob_middle*sd.max(0,(households_connected-tv_ownership_middle))
tv_ownership_middle.equation = tv_adoption_middle

#stove adoption equations
stove_social_influence_middle.equation = (stove_ownership_middle/no_of_households)*beta_social_stove
stove_utility_middle.equation = (beta_income_stove*income) + (beta_reliability_stove*reliability) + (stove_social_influence_middle)
stove_adoption_prob_middle.equation = 1 / (1 + sd.exp(-(alpha_stove_middle+(stove_utility_sensitivity*stove_utility_middle))))
stove_adoption_middle.equation = stove_adoption_prob_middle*sd.max(0,(households_connected-stove_ownership_middle))
stove_ownership_middle.equation = stove_adoption_middle

#lighting adoption equations
lighting_social_influence_middle.equation = (lighting_ownership_middle/no_of_households)*beta_social_lighting
lighting_utility_middle.equation = (beta_income_lighting*income) + (beta_reliability_lighting*reliability) + (lighting_social_influence_middle)
lighting_adoption_prob_middle.equation = 1 / (1 + sd.exp(-(alpha_lighting_middle+(lighting_utility_sensitivity*lighting_utility_middle))))
lighting_adoption_middle.equation = lighting_adoption_prob_middle*sd.max(0,(households_connected-lighting_ownership_middle))
lighting_ownership_middle.equation = lighting_adoption_middle

social_influence_microgrid.equation = households_connected/no_of_households
household_microgrid_adoption.equation = households_not_connected * ((1e-2 * reliability) + (0.5e-2*social_influence_microgrid)+(5e-4*health_improvements)+(1e-2*cost_savings_overall))
households_connected.equation = household_microgrid_adoption
households_not_connected.equation = -household_microgrid_adoption
baseline_demand.equation = baseline_household_demand * households_connected

tv_demand.equation = (tv_ownership_middle)*0.1*6*30 #kWh times hours used(hours*kWh rating*days)
stove_demand.equation = (stove_ownership_middle)*5*2.5*30 #kWh times hours used(hours*kWh rating*days)
lighting_demand.equation = (lighting_ownership_middle)*5*5*0.06*30 #kWh times hours used(bulbs*hours*kWh rating*days)

total_demand.equation = baseline_demand+tv_demand+stove_demand+lighting_demand
power_use.equation = sd.min(total_demand,microgrid_capacity)
microgrid_income_flow.equation = power_use * cost_per_kwh
net_microgrid_income.equation = microgrid_income_flow
mttr.equation = sd.lookup(
    net_microgrid_income,
    [(0, 48), (30000000, 42), (70000000, 36), (100000000, 30)]
)

failure_rate_multiplier.equation = sd.lookup(
    power_use,
    [(90000, 1), (100000, 3), (115000, 6), (130000, 8)]
)
failure_rate.equation = initial_failure_rate*no_of_components*failure_rate_multiplier
no_of_failures.equation = failure_rate
fixture_rate.equation = no_of_current_repairs/mttr
#number of components stock
no_of_components.equation = fixture_rate - failure_rate
#number of components being repaired
no_of_current_repairs.equation = failure_rate - fixture_rate
# Failure rate depends on maintenance delay
downtime.equation = mttr * no_of_failures
# number of customers
no_of_customers.equation = households_connected*5
# SAIDI (interruptions per customer)
saidi.equation = downtime / sd.max(1, no_of_customers)
# SAIFI (interruptions per customer)
saifi.equation = no_of_failures / sd.max(1, no_of_customers)
# Reliability improves as SAIDI decreases
reliability.equation = 1/(1+7*saidi)

#fossil fuel use
fossil_fuel_baseline_cooking.equation = no_of_households*0.5*1*30 #no of stoves*number of hours*fuel use per hour*number of days
fossil_fuel_baseline_lighting.equation = no_of_households*5*0.5*30 #no of lighting*number of hours*fuel use per hour*number of days

fossil_fuel_use_cooking.equation = (no_of_households-stove_ownership_middle)*0.5*1*30
fossil_fuel_use_lighting.equation = (no_of_households-lighting_ownership_middle)*5*0.5*30

#CO2/PM 2.5 emissions
# CO2 emissions from fossil fuel use
co2_emissions_kerosene.equation = fossil_fuel_use_lighting*co2_emissions_factor_kerosene
co2_emissions_lpg.equation = fossil_fuel_use_cooking*co2_emissions_factor_lpg

pm25_emissions_kerosene.equation = 0.5*1*30*pm25_emissions_factor_kerosene
pm25_emissions_lpg.equation = 0.5*1*30*pm25_emissions_factor_lpg

#PM2.5 concentration
pm25_concentration_daily_kerosene.equation = ((pm25_emissions_kerosene/30)*fraction_of_emissions_kerosene)/((air_change_rate*24)*kitchen_volume)
#daily average pm 2.5 concentration (mass of emissions per day * fraction entering kitchen)/(air changes per day*kitchen volume)
pm25_concentration_daily_lpg.equation = ((pm25_emissions_lpg/30)*fraction_of_emissions_lpg)/((air_change_rate*24)*kitchen_volume)
pm25_concentration_daily_both.equation = pm25_concentration_daily_lpg+pm25_concentration_daily_kerosene

#Cases
pm25_exposure_kerosene_men.equation = 0.450*pm25_concentration_daily_kerosene*1e6
pm25_exposure_kerosene_women.equation = 0.628*pm25_concentration_daily_kerosene*1e6
pm25_exposure_kerosene_children.equation = 0.742*pm25_concentration_daily_kerosene*1e6

pm25_exposure_lpg_men.equation = 0.450*pm25_concentration_daily_lpg*1e6
pm25_exposure_lpg_women.equation = 0.628*pm25_concentration_daily_lpg*1e6
pm25_exposure_lpg_children.equation = 0.742*pm25_concentration_daily_lpg*1e6

pm25_exposure_both_men.equation = 0.450*pm25_concentration_daily_both*1e6
pm25_exposure_both_women.equation = 0.628*pm25_concentration_daily_both*1e6
pm25_exposure_both_children.equation = 0.742*pm25_concentration_daily_both*1e6


prob_kerosene.equation = 1-(lighting_ownership_middle/no_of_households)
prob_lpg.equation = 1-(stove_ownership_middle/no_of_households)
prob_both.equation = (prob_lpg*prob_kerosene)

estimated_cases_kerosene_men.equation = ((no_of_households-lighting_ownership_middle)*1)*incident_rate*(1-(1/(1+(0.3*(1-(sd.exp(-1.2*(pm25_exposure_kerosene_men-2.5)**1.2)))))))#IER function
estimated_cases_kerosene_women.equation = ((no_of_households-lighting_ownership_middle)*2)*incident_rate*(1-(1/(1+(0.3*(1-(sd.exp(-1.2*(pm25_exposure_kerosene_women-2.5)**1.2)))))))
estimated_cases_kerosene_children.equation = ((no_of_households-lighting_ownership_middle)*2)*incident_rate*(1-(1/(1+(0.3*(1-(sd.exp(-1.2*(pm25_exposure_kerosene_children-2.5)**1.2)))))))

estimated_cases_lpg_men.equation = ((no_of_households-stove_ownership_middle)*1)*incident_rate*(1-(1/(1+(0.3*(1-(sd.exp(-1.2*(pm25_exposure_lpg_men-2.5)**1.2)))))))
estimated_cases_lpg_women.equation = ((no_of_households-stove_ownership_middle)*2)*incident_rate*(1-(1/(1+(0.3*(1-(sd.exp(-1.2*(pm25_exposure_lpg_women-2.5)**1.2)))))))
estimated_cases_lpg_children.equation = ((no_of_households-stove_ownership_middle)*2)*incident_rate*(1-(1/(1+(0.3*(1-(sd.exp(-1.2*(pm25_exposure_lpg_children-2.5)**1.2)))))))

estimated_cases_both_men.equation = (no_of_households*prob_both*1)*incident_rate*(1-(1/(1+(0.3*(1-(sd.exp(-1.2*(pm25_exposure_both_men-2.5)**1.2)))))))
estimated_cases_both_women.equation = (no_of_households*prob_both*2)*incident_rate*(1-(1/(1+(0.3*(1-(sd.exp(-1.2*(pm25_exposure_both_women-2.5)**1.2)))))))
estimated_cases_both_children.equation = (no_of_households*prob_both*2)*incident_rate*(1-(1/(1+(0.3*(1-(sd.exp(-1.2*(pm25_exposure_both_children-2.5)**1.2)))))))
# assuming each household has 1 man 2 women and 2 children

estimated_cases_men.equation = estimated_cases_kerosene_men + estimated_cases_lpg_men - estimated_cases_both_men
estimated_cases_women.equation = estimated_cases_kerosene_women + estimated_cases_lpg_women - estimated_cases_both_women
estimated_cases_children.equation = estimated_cases_kerosene_children + estimated_cases_lpg_children - estimated_cases_both_children

estimated_cases_total.equation = estimated_cases_men + estimated_cases_women + estimated_cases_children


#baseline cases
baseline_cases_both_men.equation = (no_of_households*1)*incident_rate*(1-(1/(1+(0.3*(1-(sd.exp(-1.2*(pm25_exposure_both_men-2.5)**1.2)))))))
baseline_cases_both_women.equation = (no_of_households*2)*incident_rate*(1-(1/(1+(0.3*(1-(sd.exp(-1.2*(pm25_exposure_both_women-2.5)**1.2)))))))
baseline_cases_both_children.equation = (no_of_households*2)*incident_rate*(1-(1/(1+(0.3*(1-(sd.exp(-1.2*(pm25_exposure_both_children-2.5)**1.2)))))))

baseline_cases_total.equation = baseline_cases_both_men+baseline_cases_both_women+baseline_cases_both_children

#Health function
health_improvements.equation = (1-estimated_cases_total/baseline_cases_total)

#savings
baseline_lpg_cost.equation = lpg_price_per_kg*0.5*1*30
stove_cost.equation = cost_per_kwh*5*2.5*30
cost_savings_lpg.equation = (baseline_lpg_cost-stove_cost)/baseline_lpg_cost

baseline_kerosene_cost.equation = kerosene_price_per_liter*0.5*1*30
lighting_cost.equation = cost_per_kwh*5*2.5*30
cost_savings_kerosene.equation = (baseline_kerosene_cost-lighting_cost)/baseline_kerosene_cost
#-(stove_cost+lighting_cost)
total_baseline.equation = (baseline_lpg_cost+baseline_kerosene_cost)
total_electric_cost.equation = (stove_cost+lighting_cost)
cost_savings_both.equation = (total_baseline-total_electric_cost)/total_baseline

prob_stove.equation = 1-prob_lpg
prob_lighting.equation = 1-prob_kerosene
prob_stove_and_lighting.equation = prob_stove*prob_lighting

cost_savings_overall.equation = ((prob_stove-prob_stove_and_lighting)*cost_savings_lpg)+((prob_lighting-prob_stove_and_lighting)*cost_savings_kerosene)+(prob_stove_and_lighting*cost_savings_both)

#-----------------------------------------------------------------
#Reinforcing Loop 4 (R4)
#-----------------------------------------------------------------
employment = model.stock("employment")
available_jobs = model.stock("available_jobs")

job_fill = model.flow("job_fill")
job_creation = model.flow("job_creation")
job_attrition = model.flow("job_attrition")

avg_hours_saved_women = model.converter("avg_hours_saved_women")
avg_hours_saved_men = model.converter("avg_hours_saved_men")
willingness_women = model.converter("willingness_women")
willingness_men = model.converter("willingness_men")
labour_supply = model.converter("labour_supply")
income = model.converter("income")
consumption = model.converter("consumption")
business_income = model.converter("business_income")
capital_investment = model.converter("capital_investment")
no_of_households = model.constant("no_of_households")
men_per_household = model.constant("men_per_household")
women_per_household = model.constant("women_per_household")
resp_elec_women = model.constant("resp_elec_women")
resp_elec_men = model.constant("resp_elec_men")
average_wage = model.constant("average_wage")
mpc = model.constant("mpc")
local_spending_fraction = model.constant("local_spending_fraction")
investment_rate = model.constant("investment_rate")
job_creation_efficiency = model.constant("job_creation_efficiency")
attrition_rate = model.constant("attrition_rate")


# Time savings from electrification
avg_hours_saved_women.equation = (households_connected/no_of_households) * 2.0
avg_hours_saved_men.equation = (households_connected/no_of_households) * 0.5

# Willingness to participate in labor market
willingness_women.equation = avg_hours_saved_women * resp_elec_women
willingness_men.equation = avg_hours_saved_men * resp_elec_men

# Aggregate labour supply (people, not hours)
labour_supply.equation = (
    willingness_men * men_per_household +
    willingness_women * women_per_household
) * no_of_households

# Job filling constrained by available jobs
job_fill.equation = sd.min(
    labour_supply,
    sd.max(0, available_jobs - employment)
)

# Employment stock
employment.equation = job_fill - job_attrition

# Household income (average per household)
income.equation = ((employment * average_wage) / no_of_households)*future_income_multiplier

# Consumption
consumption.equation = mpc * income

# Business income
business_income.equation = local_spending_fraction * consumption

# Capital investment
capital_investment.equation = investment_rate * business_income

# Job creation and destruction
job_creation.equation = job_creation_efficiency * capital_investment
job_attrition.equation = attrition_rate * employment

# Available jobs stock
available_jobs.equation = job_creation - job_attrition

employment.initial_value = 100.0
available_jobs.initial_value = 200.0

men_per_household.equation = 1.0
women_per_household.equation = 2.0

resp_elec_women.equation = 0.25
resp_elec_men.equation = 0.15

average_wage.equation = 1000
mpc.equation = 0.75
local_spending_fraction.equation = 0.6
investment_rate.equation = 0.25

job_creation_efficiency.equation = 0.05
attrition_rate.equation = 0.02

#-----------------------------------------------------------------
#Reinforcing Loop 5 (R5)
#-----------------------------------------------------------------
#*****************************************************************
human_capital = model.stock("human_capital")

learning = model.flow("learning")

study_time = model.converter("study_time")
future_income_multiplier = model.converter("future_income_multiplier")

study_time_sensitivity = model.constant("study_time_sensitivity")
learning_efficiency = model.constant("learning_efficiency")
human_capital_decay = model.constant("human_capital_decay")
returns_to_education = model.constant("returns_to_education")

# Electrification increases study time
study_time.equation = study_time_sensitivity * (households_connected/no_of_households)

# Human capital accumulation
learning.equation = learning_efficiency * study_time
human_capital.equation = learning - human_capital_decay * human_capital

# Education affects future income
future_income_multiplier.equation = 1 + returns_to_education * human_capital

human_capital.initial_value = 1.0

study_time_sensitivity.equation = 2.0
learning_efficiency.equation = 0.05
human_capital_decay.equation = 0.01
returns_to_education.equation = 0.08

#*****************************************************************


#initialize model
tv_ownership_middle.initial_value=0.0
baseline_household_demand.equation = 200.0 #in kWh
tv_utility_sensitivity.equation = 2.45
initial_tv_composition.equation = 30.0
alpha_tv_middle.equation = -6.0
beta_income_tv.equation = 1e-5
beta_reliability_tv.equation = 0.5
beta_social_tv.equation = 0.2

stove_ownership_middle.initial_value=0.0
alpha_stove_middle.equation = -9.9
beta_income_stove.equation = 1e-3
beta_reliability_stove.equation = 0.4
beta_social_stove.equation = 0.25
initial_stove_composition.equation = 30
stove_utility_sensitivity.equation = 1.0

lighting_ownership_middle.initial_value=0.0
alpha_lighting_middle.equation = -0.1
beta_income_lighting.equation = 1e-10
beta_reliability_lighting.equation = 0.6
beta_social_lighting.equation = 0.1
initial_lighting_composition.equation = 0.0
lighting_utility_sensitivity.equation = 0.7

#income.equation = 6000.0
no_of_households.equation = 600.0
households_connected.initial_value = 0.0 
households_not_connected.initial_value = no_of_households
microgrid_capacity.equation = 1000000.0 #KW
cost_per_kwh.equation = 0.18
initial_failure_rate.equation = (6e-6)*720.0
no_of_failures.initial_value = 0.0
no_of_components_initial_value.equation = 2.0
no_of_components.initial_value = no_of_components_initial_value
no_of_current_repairs.initial_value = 0.0

co2_emissions_factor_kerosene.equation = 2.7
co2_emissions_factor_lpg.equation = 5
pm25_emissions_factor_kerosene.equation = 10 #g/L
pm25_emissions_factor_lpg.equation = 5
fraction_of_emissions_kerosene.equation = 0.25
fraction_of_emissions_lpg.equation = 0.4
air_change_rate.equation = 2
kitchen_volume.equation = 20
incident_rate.equation = 7e-4

lpg_price_per_kg.equation = 10
kerosene_price_per_liter.equation = 15

#cost_savings_lpg.plot()
#cost_savings_kerosene.plot()
#cost_savings_both.plot()
#cost_savings_overall.plot()
#baseline_kerosene_cost.plot()
#lighting_cost.plot()

print(cost_savings_both(0))
#health_improvements.plot()

#prob_kerosene.plot()
#prob_lpg.plot()
#prob_both.plot()

#estimated_cases_both_men.plot()
#estimated_cases_both_women.plot()
#estimated_cases_both_children.plot()

#estimated_cases_women.plot()
#estimated_cases_children.plot()
#estimated_cases_total.plot()
#baseline_cases_total.plot()
#print(estimated_cases_total(0))
#print(baseline_cases_total(0))
#human_capital.plot()
#avg_hours_saved_women.plot()
#willingness_women.plot()
#labour_supply.plot()
#employment.plot()
#income.plot()
#future_income_multiplier.plot()
#consumption.plot()
#business_income.plot()
#capital_investment.plot()
#available_jobs.plot()
#fossil_fuel_use_cooking.plot()
#fossil_fuel_use_lighting.plot()
#households_connected.plot()
#tv_ownership_middle.plot()
#stove_ownership_middle.plot()
#lighting_ownership_middle.plot()
#tv_demand.plot()
#stove_demand.plot()
#lighting_demand.plot()
#print(total_demand(100))
households_connected.plot()
#households_not_connected.plot()
#total_demand.plot()
#net_microgrid_income.plot()
#mttr.plot()
#total_demand.plot()
#power_use.plot()
#no_of_failures.plot()
#downtime.plot()
#no_of_customers.plot()
#saidi.plot()
#saifi.plot()
#reliability.plot()
```
